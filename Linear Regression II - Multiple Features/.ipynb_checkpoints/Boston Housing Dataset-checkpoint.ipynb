```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()

X = boston.data
y = boston.target
```

```python
print(X.shape,y.shape)
```

```python
print(boston.feature_names)
```

```python
print(boston.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(X,columns=boston.feature_names)
df.head()
```

```python
df.describe()
```

### From here we see that the data is not normalised
- we need to normalise it

```python
import numpy as np
```

```python
u = np.mean(X,axis=0)
std = np.std(X,axis=0)
```

```python
X = (X-u)/std
```

```python
df = pd.DataFrame(X)
```

```python
df.head()
```
