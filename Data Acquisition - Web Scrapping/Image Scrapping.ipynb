```python
import bs4
import requests
```

```python
url = "https://www.values.com/inspirational-quotes"
```

```python
response = requests.get(url)
```

```python
response.content
```

```python
soup = bs4.BeautifulSoup(response.content)
```

```python
print(soup)
```

```python
article_elements=soup.findAll('div',{"class":"col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top"})
```

```python
len(article_elements)
```

```python
article_elements[0]
```

```python
img_src = [x.findAll('img') for x in article_elements]    
```

```python
len(img_src)
```

```python
for src in img_src:
    print(src[0].attrs["src"])
```

```python
for idx,src in enumerate(img_src):
    curr_img = requests.get(src[0].attrs['src'])
    
    with open(f"image{idx}.jpg","wb") as f:
        f.write(curr_img.content)
```
