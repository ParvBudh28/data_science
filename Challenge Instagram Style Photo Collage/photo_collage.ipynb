```python
import cv2
import numpy as np
```

```python
top_left = cv2.imread('./ImageCollage dataset/top_left.jpg')
top_right = cv2.imread('./ImageCollage dataset/top_right.jpg')
```

```python
bottom_left = cv2.imread('./ImageCollage dataset/bottom_left.jpg')
bottom_right = cv2.imread('./ImageCollage dataset/bottom_right.jpg')
```

```python
center = cv2.imread('./ImageCollage dataset/center.jpeg')
```

```python
top_left = cv2.resize(top_left,(200,200))
```

```python
top_right = cv2.resize(top_right,(200,200))
bottom_left = cv2.resize(bottom_left,(200,200))
bottom_right = cv2.resize(bottom_right,(200,200))
```

```python
center = cv2.resize(center,(100,100))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(cv2.cvtColor(center,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB))
```

```python
print(type(top_left))
```

```python
final_image = np.zeros((430,430,3),dtype=np.uint8)
```

```python
plt.imshow(final_image)
```

```python
final_image[10:210,10:210] = cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(final_image)
```

```python
final_image[10:210,220:420] = cv2.cvtColor(top_right,cv2.COLOR_BGR2RGB)
```

```python
final_image[220:420,10:210] = cv2.cvtColor(bottom_left,cv2.COLOR_BGR2RGB)
final_image[220:420,220:420] = cv2.cvtColor(bottom_right,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(final_image)
```

```python
plt.imshow(center_image)
```
