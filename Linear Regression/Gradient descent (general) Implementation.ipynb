# Gradient Descent implementation(general)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

### given a function f, we want to find value of x that minimises f(x)

## Visualisation

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = 0
lr = 0.1 #step size
error = []

# take 50 steps in the downhill direction
for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    print(x)
    
    error.append((x-5)**2)
    plt.scatter(x,error[-1])
```

```python
# plot the values of error
plt.plot(error)
```
