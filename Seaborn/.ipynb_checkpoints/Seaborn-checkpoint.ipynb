# Seaborn
[seaborn documentation](https://seaborn.pydata.org/introduction.html#:~:text=Seaborn%20is%20a%20library%20for,explore%20and%20understand%20your%20data.)

`pip install seaborn`
- built on top of matplotlib
- advanced data visualisation techniques
- works well with pandas data frames

```python
import seaborn as sns
import numpy as np
```

```python
# loads an example data set
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
print(type(tips))
```

```python
sns.barplot(x='sex',y='total_bill', data=tips)
```

```python
sns.barplot(x='sex',y='total_bill', data=tips,estimator=np.std)
```

```python
# here y-axis by default is count
sns.countplot(x = 'sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x='day' ,y='total_bill' ,data=tips)
```

```python
sns.boxplot(x='day' ,y='total_bill' ,data=tips,hue='smoker')
```

```python
sns.boxplot(x='day' ,y='total_bill' ,data=tips,hue='sex')
```

```python
# outliers in box plots
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

## Distribution Plots

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
tips_corr = tips.corr()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

- visualise correlation among survived pclass sex age
- come up with interesting conclusions and do analysis of probability of surviving with a particular sex and class

```python
only_males = titanic.loc[titanic['sex'].isin(["male"])]
males_survived = sns.countplot(x='survived',data=only_males,hue='class')
```

```python
only_females = titanic.loc[titanic['sex'].isin(["female"])]
females_survived = sns.countplot(x='survived',data=only_females,hue='class')
```

```python
type(titanic)
```

```python
print(len(only_females))
first_class_females = only_females.loc[only_females['pclass'].isin([1])]
print(len(first_class_females))
second_class_females = only_females.loc[only_females['pclass'].isin([2])]
print(len(second_class_females))
third_class_females = only_females.loc[only_females['pclass'].isin([3])]
print(len(third_class_females))
```
