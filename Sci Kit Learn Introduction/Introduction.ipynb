```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=1000,n_features=2,n_informative=2,noise=10,random_state=1)
```

```python
print(X.shape,y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)

plt.show()
```

```python
from mpl_toolkits import mplot3d
%matplotlib notebook   # to make use of rotation of 3d plots
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')

axes.scatter3D(X[:,0],X[:,1],y,color='green')
plt.title('3d scatter plot')
plt.show()
```

## Implementing Linear Regression Using Sk-Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
# theta 1....n
model.coef_
```

```python
# theta 0
model.intercept_
```

```python
# make predictions
model.predict([X[0],X[1]])
```

```python
# actual values
y[0],y[1]
```

```python
# Coeffecient of determination
model.score(X,y)
```
