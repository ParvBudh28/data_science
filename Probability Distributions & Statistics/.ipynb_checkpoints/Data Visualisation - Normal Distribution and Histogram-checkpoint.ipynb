## Generating and plotting (visualising) Normal Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u=5
sigma=1

vals = np.random.randn(1000)
vals = u+sigma*vals

plt.hist(vals,100)
plt.show()
```

```python
plt.hist(vals,10)
plt.show()
```

```python
vals = np.rint(vals)
```

```python
np.unique(vals,return_counts=True)
```

```python
x = vals[:200]
y = np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

#### Density is more near the mean

## Multivariable Normal Distribution

```python
# Bivariate
mean = np.array([0,0])
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean,cov,500)
# print(dist)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.show()

# here sigma x,y is 0 therefore there is no correlation between x and y
```

```python
# Bivariate
mean = np.array([0,0])
cov = np.array([[1,0.8],[0.8,1]])

mean2 = np.array([5,6])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist = np.random.multivariate_normal(mean,cov,500)
# print(dist)
print(dist.shape)
dist2 = np.random.multivariate_normal(mean2,cov2,500)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()

# here sigma x,y is 0 therefore there is no correlation between x and y
```
