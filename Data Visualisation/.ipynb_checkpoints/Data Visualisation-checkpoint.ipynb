## Line Plots
`plt.plot()`

```python
# dummy data
x = np.arange(10)
y1 = x**2
y2 = 2*x+3
```

```python
#plt.plot?
```

```python
plt.plot(x,y1,color='red',label='Apple',marker='o')
plt.plot(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')

# to provide axes labels
plt.xlabel("Time")
plt.ylabel("Price")

# to give title of the graph
plt.title("Prices of fruits over Time")

# To show the descriptors shown on top left
plt.legend()

plt.show()   # to print the graph on screen

# whenever we call plt.show() method it means that we have finished the graph
```

```python
plt.plot(x,y1,color='red')
plt.show()
plt.plot(x,y2,color='green')
plt.show()
```

```python
prices = np.array([1,2,3,4])**3
print(prices)
```

```python
# here the indices are on x-axis and prices go on the y-axis if we pass only one array
plt.plot(prices,color='red')
plt.show()
```

## Scatter Plots
`plt.scatter()`

```python
# adjusting the size of any plots
plt.figure(figsize=(6,6))

plt.scatter(x,y1,color='red',label='Apple',marker='^')
plt.scatter(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')

# to provide axes labels
plt.xlabel("Time")
plt.ylabel("Price")

# to give title of the graph
plt.title("Prices of fruits over Time")

# To show the descriptors shown on top left
plt.legend()

plt.show()   # to print the graph on screen

# whenever we call plt.show() method it means that we have finished the graph
```

## Bar - graphs

```python
# plt.bar?
```

```python
plt.style.use("seaborn-dark")

x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,30],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"]) #Current year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year",color="orange") #Next year
plt.legend()
plt.title("Metal Price Comparison")

plt.ylim(0,50)

plt.xlabel("Metal")
plt.ylabel("Price")
plt.show()
```

```python
plt.style.use("dark_background")

x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,30],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"]) #Current year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year") #Next year
plt.legend()
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.show()
```

## Pie Chart

```python
# plt.pie?
```

```python
subjects = "Maths","Chem","Physics","English"
weightage = [20,10,15,5]

plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),autopct='%1.1f%%')

plt.show()
```

## Histogram

```python
# generate data using standard normal
```

```python
xsn = np.random.randn(100)
sigma = 5
u = 60
x = np.round(xsn*sigma + u)
x2 = np.round(xsn*10+40)
print(x)
```

```python
plt.style.use("seaborn")
plt.hist(x,alpha=0.9,label="Maths")
plt.hist(x2,alpha=0.8,label="Physics")
plt.legend()
plt.title("Histogram")
plt.ylabel("Freq count of students")
plt.xlabel("Marks range")
plt.show()
```
