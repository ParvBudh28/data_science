## Displaying and working with images only using Open CV

```python
import cv2
# "Cat Image" is the title
img = cv2.imread("cat.jpg")
cv2.imshow("Cat Image",img)
gray_scale = cv2.imread("cat.jpg",cv2.IMREAD_GRAYSCALE)
cv2.imshow("Gray Cat",gray_scale)
# waitKey(0) tells to wait for infinte time to display the window until program stopped
# waitKey(x) waits for x ms to close the window
cv2.waitKey(2500)

# destroy all windows after pausing the program
cv2.destroyAllWindows()
```

### Matplotlib interprets as RGB wherease cv2 itself interprets as BGR . therefore no need to convert
